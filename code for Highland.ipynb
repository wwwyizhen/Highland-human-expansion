# Fig. 5a

```python
import numpy as np
import pandas as pd
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error

# =======================
# 读取数据
# =======================
df = pd.read_excel(r"delta_change.xlsx", sheet_name='Sheet3')

# 自变量 X （可以选择多个）
X = df[["delta_frg", "delta_lst",  "delta_gpp","delta_pre", "delta_pdsi", "delta_sm","delta_ws","delta_fvc"]].to_numpy()  # LV1 和 LV2 对应的指标
y = df["delta_thr"].to_numpy()  

# =======================
# 训练 XGBoost 回归模型
# =======================
model_xgb = XGBRegressor()

model_xgb.fit(X, y)


y_pred = model_xgb.predict(X)

r2 = r2_score(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

print(f"R² = {r2:.3f}")
print(f"RMSE = {rmse:.3f}")


import matplotlib.pyplot as plt
import numpy as np

features = ["ΔMPS", "ΔLST", "ΔGPP", "ΔP", "ΔPDSI", "ΔSM", "ΔWS", "ΔFVC"]
importances = model_xgb.feature_importances_
indices = np.argsort(importances)[::-1]


fig, ax = plt.subplots(1, figsize=(3.5, 3.5), dpi=800, facecolor='white', edgecolor='white', frameon=True)
fig.patch.set_alpha(0)
ax.axvspan(-0.5, 1.5, color='lightgray', alpha=1) 
plt.xlim([-0.5,7.5])
plt.bar(range(len(features)), importances[indices], align="center", color="C0")
plt.xticks(range(len(features)), np.array(features)[indices], rotation=45)
plt.xlabel("Potential factors")
plt.ylabel("Feature Importance")
plt.tight_layout()
plt.show()
```

# Fig. 5b

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

df = pd.read_excel(r"delta_change.xlsx", sheet_name='Sheet3')

x = df["delta_frg"].values     
y = df["delta_lst"].values     
z = df["delta_thr"].values   
xi = np.linspace(-200, 200, 10)
yi = np.linspace(-0.5, 0.5,10)

X, Y = np.meshgrid(xi, yi)


Z = griddata((x, y), z, (X, Y), method='cubic')
fig, ax = plt.subplots(figsize=(3.5,3.5), dpi=600, facecolor='white')
fig.patch.set_alpha(0)

contour = ax.contourf(X, Y, Z, levels=100, cmap="Spectral_r")
cbar = plt.colorbar(contour, ax=ax)

plt.grid(True, linestyle='--', linewidth=0.2,alpha=0.7)
# 设置坐标轴标签
ax.set_ylabel("ΔLST (°C)")
ax.set_xlabel("ΔMPS (ha)")
cbar.set_label("ΔMSA")

plt.show()
```
